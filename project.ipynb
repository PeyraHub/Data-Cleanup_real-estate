#### Exercise 09. Are the average prices of "Valdemorillo" and "Galapagar" the same? (★★☆)

Print both average prices and then write a conclusion about them.

```python
import pandas as pd
ds = pd.read_csv('assets/real_estate.csv', sep=';')

def average_price(population):
    filter_population = ds[ds['level5'] == population]
    average_price = filter_population['price'].mean() 
    return average_price

valdemorillo_average_price = average_price("Valdemorillo")
galapagar_average_price = average_price("Galapagar")

print(f"The average price in Valdemorillo is {valdemorillo_average_price}")
print(f"The average price in Galapagar is {galapagar_average_price}")

if valdemorillo_average_price == galapagar_average_price:
    print("The average price in Valdemorillo and Galapagar is the same.")

else:
    print("The average price in Valdemorillo and Galapagar is not the same.")
```

#### Exercise 10. Are the average prices per square meter (price/m2) of "Valdemorillo" and "Galapagar" the same? (★★☆)

Print both average prices and then write a conclusion about it.

Hint: Create a new column called `pps` (price per square meter) and then analyze the values.

```python
import pandas as pd
ds = pd.read_csv('assets/real_estate.csv', sep=';')

ds['pps'] = ds['price'] / ds['surface']

def average_pps(population):
    filter_population = ds[ds['level5'] == population]
    average_pps = filter_population['pps'].mean()  # Calcula el precio promedio
    return average_pps

average_pps_valdemorillo = average_pps("Valdemorillo")
average_pps_galapagar = average_pps("Galapagar")

print(f"The average price per square meter in Valdemorillo is {average_pps_valdemorillo}")
print(f"The average price per square meter in Galapagar is {average_pps_galapagar}")

if average_pps_valdemorillo == average_pps_galapagar:
    print("The average price per square meter in Valdemorillo and Galapagar is the same.")

else:
    print("The average price per square meter in Valdemorillo and Galapagar is not the same.")
```

#### Exercise 19. What can you say about the price per square meter (price/m2) between the towns of "Getafe" and "Alcorcón"? You must use the subset obtained in Exercise 14 (★★☆)

Hint: Create a new column called `pps` (price per square meter) and then analyze the values.

```python
import pandas as pd
ds = pd.read_csv('assets/real_estate.csv', sep=';')

ds['pps'] = ds['price'] / ds['surface']

def get_average(population_name):
    filtered_ds = ds[ds['level5'] == population_name]
    average_pps = filtered_ds['pps'].mean()  # Calcula el promedio de 'pps' (precio/m²)
    return average_pps

average_pps_getafe = get_average("Getafe")
average_pps_alcorcon = get_average("Alcorcón")

print(f"Average price per square meter in Getafe: {average_pps_getafe}")
print(f"Average price per square meter in Alcorcon: {average_pps_alcorcon}")

if average_pps_getafe == average_pps_alcorcon:
    print("Average prices per square meter in Getafe and Galapagar are equal.")
else:
    print("Average prices per square meter in Getafe and Galapagar are not equal.")
```

#### Exercise 20. Make the same plot for 4 different populations (level5 column) and rearrange them on the same graph. You must use the subset obtained in Exercise 14 (★★☆)
 
Hint: Make a scatter plot of each population using subplots.

```python
import pandas as pd
import matplotlib.pyplot as plt

ds = pd.read_csv('assets/real_estate.csv', sep=';')

south_belt_populations = ["Fuenlabrada", "Leganés", "Getafe", "Alcorcón"]
south_belt_subset = ds[ds['level5'].isin(south_belt_populations)]

fig, axs = plt.subplots(2, 2, figsize=(10, 10))

subplot_positions = {
    "Fuenlabrada": (0, 0),
    "Leganés": (0, 1),
    "Getafe": (1, 0),
    "Alcorcón": (1, 1)
}

for population, pos in subplot_positions.items():
    population_data = south_belt_subset[south_belt_subset['level5'] == population]
    
    axs[pos].scatter(population_data['surface'], population_data['price'], alpha=0.5)
    axs[pos].set_title(f"{population}")
    axs[pos].set_xlabel("Superficie (m2)")
    axs[pos].set_ylabel("Precio (€)")

plt.tight_layout()
plt.show()
```

```python
import pandas as pd

ds = pd.read_csv('assets/real_estate.csv', sep=';')
south_belt_populations = ["Fuenlabrada", "Leganés", "Getafe", "Alcorcón"]
south_belt_subset = ds[ds['level5'].isin(south_belt_populations)]

coordinates_by_population = {
    population: list(zip(south_belt_subset[south_belt_subset['level5'] == population]['latitude'],
                         south_belt_subset[south_belt_subset['level5'] == population]['longitude']))
    for population in south_belt_populations
}

print(coordinates_by_population)
```

```python
from ipyleaflet import Map, Marker, Icon, basemaps

population_colors = {
    'Fuenlabrada': 'red',
    'Leganés': 'blue',
    'Getafe': 'green',
    'Alcorcón': 'orange'
}

madrid_center = (40.4168, -3.7038)  # Approximate center (pinned from google maps not from ds)
map = Map(center=madrid_center, zoom=10, basemap=basemaps.OpenStreetMap.Mapnik)

for population, coords in coordinates_by_population.items():
    latitudes = [float(str(coord[0]).replace(',', '.')) for coord in coords]  # Convert latitudes to float
    longitudes = [float(str(coord[1]).replace(',', '.')) for coord in coords]  # Convert longitudes to float
    
    color = population_colors[population]
    for lat, lon in zip(latitudes, longitudes):
        icon = Icon(icon_url=f"http://maps.google.com/mapfiles/ms/icons/{color}-dot.png")
        marker = Marker(location=(lat, lon), title=population, icon=icon, draggable=False)
        map.add_layer(marker)  # Add the marker to the map

map
```
